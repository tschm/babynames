```python
import pandas as pd
import numpy as np
%matplotlib inline  
pd.options.display.max_rows=20
```

```python
from os.path import expanduser, join, exists
home = expanduser("~")
boys = pd.read_csv(join(home, "work", "data", "ch", "chMale.csv"), index_col=0).truncate(before=1910)
girls = pd.read_csv(join(home, "work", "data", "ch", "chFemale.csv"), index_col=0).truncate(before=1910)
```

## Each name lives on the unit-simplex

## Each name can be projected to the unit-sphere

## We compute the Bhattacharyya angle (correlation)

```python
# project to unit-simplex
def proj_simplex(ts):
    return ts.fillna(0)/ts.sum()
    
# project to unit-sphere 
def proj_sphere(ts):
    return np.sqrt(proj_simplex(ts))
```

```python
def match(body1, body2):
    if isinstance(body1, pd.Series):
        body1 = body1.to_frame()
    if isinstance(body2, pd.Series):
        body2 = body2.to_frame()
    x = body1.apply(proj_sphere)
    y = body2.apply(proj_sphere)
    return pd.DataFrame({key : x[key].dot(y) for key in x.keys()}).stack().sort_values()
```

```python
match(boys["Thomas"], girls)
```

```python
match(girls, boys["Thomas"])
```

```python
match(boys, boys["Urs"])
```

```python
match(boys, girls)
# Ruth is amazing!
```

```python
match(girls["Ruth"], boys)
```

```python
def pair_plot(ts1, ts2):
    a = proj_simplex(ts1)
    b = proj_simplex(ts2)
    print(match(ts1, ts2))
    pd.concat((a,b), axis=1).plot()
```

```python
pair_plot(boys["Thomas"], girls["Lydia"])
```

```python
pair_plot(boys["Chris"], girls["Rachel"])
```

```python
pair_plot(boys["Urs"], boys["Beat"])
```

```python
pair_plot(boys["Walter"], girls["Ruth"])
```
