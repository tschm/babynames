# The Elvis effect

```python
import pandas as pd
%matplotlib inline
pd.options.display.max_rows=20
```

```python
names = pd.read_csv("us.csv", index_col=["year","Name","Gender"])["n"]
names
```

```python
f = names.loc[:,"Elvis","M"]
f.plot()
f.sort_values()
```

```python
f = names.loc[:,"Elvis","F"]
f.plot()
f.sort_values()
```

```python
f = names.loc[:,"Nikita","F"]
f.plot()
f.sort_values()
```

```python
a=names.groupby(["Name","Gender"]).sum()
a.sort_values()
```
